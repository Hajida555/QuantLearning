```python
staes = [" Alabama ", "Georgia!", "Georgia", "georgia", "FIOrIda", "south  carolian##", "West virginia?"] 
```

```python
import re
```

```python
def clean_strings(strings):
    result = []
    for value in strings:
        value = value.strip()
        value = re.sub("[!#?]","", value)
        value = value.title()
        result.append(value)
    return result
```

```python
clean_strings(staes)
```

```python
states = [" Alabama ", "Georgia!", "Georgia", "georgia", "FIOrIda", "south  carolian##", "West virginia?"] 
```

```python
def remove_punctuation(value):
    return re.sub("[!#?]", "", value)
```

```python
clean_ops = [str.strip, remove_punctuation, str.title]
```

```python
def clean_strings2(strings2, ops2):
    result2 = []
    for value2 in strings2:
        for func2 in ops2:
            value2 = func2(value2)
        result2.append(value2)
    return result2
```

```python
clean_strings2(states, clean_ops)
```

```python
for x in map(remove_punctuation, states):
    print(x)
```

```python
def square(n=10):
    print(f"Generating squares from 1 to {n ** 2}")
    for i in range(1, n + 1):
        yield i ** 2
```

```python
gen = square()
```

```python
gen
```

```python
for x in gen:
    print(x,end=" ")
```
