```python
import pandas as pd
```

```python
import numpy as np
```

```python


frame = pd.DataFrame(np.random.standard_normal((4,3)),
                     columns=list("bde"),
                     index=["Utah", "Ohio", "Texas", "Oregon"])
```

```python
frame
```

```python
np.abs(frame)
```

```python
def f1(x):
    return x.max() - x.min()
```

```python
frame.apply(f1)
```

```python
frame.apply(f1)
```

```python
frame.apply(f1, axis="columns")
```

```python
def f2(x):
    return pd.Series([x.min(), x.max()], index=["min", "max"])
```

```python
frame.apply(f2)
```

```python
def my_fotmat(x):
    return f"{x:.2f}"
```

```python
frame.map(my_fotmat)
```

```python
frame["e"].map(my_fotmat)
```
