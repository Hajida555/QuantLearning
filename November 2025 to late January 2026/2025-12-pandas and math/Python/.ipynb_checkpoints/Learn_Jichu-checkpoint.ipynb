```python
staes = [" Alabama ", "Georgia!", "Georgia", "georgia", "FIOrIda", "south  carolian##", "West virginia?"] 
```

```python
import re
```

```python
def clean_strings(strings):
    result = []
    for value in strings:
        value = value.strip()
        value = re.sub("[!#?]","", value)
        value = value.title()
        result.append(value)
    return result
```

```python
clean_strings(staes)
```
